## Regression Analysis

#### First try: GDP by Year and Medals Won

We are going to check if there is a relationship between the GDP of your country and the change of getting a medal

```python
# we are going to use the dataset of tally per year
regression = tally_per_year.copy()

# visualize the relationship between the features and the response using scatterplots
regression.plot(kind='scatter', x='GDP', y='Medal_Won', color='purple')
plt.show()
```

```python
# There is not clear relationship! Let's check this and try to figure out why
```

```python
from scipy.stats import ttest_1samp
```

```python
Medals_Mean = np.mean(regression['Medal_Won'])
Medals_Mean
```

```python
tset, pval = ttest_1samp(regression['Medal_Won'], 2.5)
```

```python
print("null hypothesis: there is NO relationship between medals won and gdp")
print("p-values",pval)
if pval < 0.05:    # alpha value is 0.05 or 5%
    print("we are rejecting null hypothesis")
else:
    print("we are failing to reject null hypothesis")
```

```python
# we confirmed the lack of relationship that the plot showed 
# but let's see what kind of model would this create...
```

```python
import statsmodels.formula.api as smf
```

```python
# create a fitted model in one line
lm = smf.ols(formula='Medal_Won ~ GDP', data=regression).fit()

# print the coefficients
lm.params
```

#### Final data frame with the principal data of Medals Won per country categorized in:
#### 1 - High Performance

```python
# find the appropriate cluster number
from sklearn.cluster import KMeans
seed = 4

inertia = []
elbow = dict()
for i in range(1, 11):
    
    estimator = KMeans(n_clusters = i,random_state=seed)
    res = estimator.fit_predict(clust2)
    inertia = estimator.inertia_
    elbow[i] = inertia
    
elbow_df = pd.Series(elbow)
ax = elbow_df.plot(title = 'Elbow Method')
ax.set_xlabel('Number of clusters')
ax.set_ylabel('Inertia')
plt.plot(2,elbow_df[2],'ro')
```

### We aplied the elbow method to find out the number of clusters that We should use for this analisys

```python
from sklearn.cluster import KMeans
model = KMeans(2)
model.fit(clust2)
print(model.cluster_centers_)
print(model.labels_)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(clust2['Medal_Won'],model.labels_)
print("Accuracy score is", acc)
```

### With the value of 2 clusters for the problem We obtain an accuracy of 0.78

```python
clusters = model.fit_predict(clust2)

# Fix permuted labels

from scipy import stats
labels = np.zeros_like(clusters)
for i in range(4):
    mask = (clusters == i)
    labels[mask] = stats.mode(clust2['Medal_Won'][mask])[0]
    
    
print("Accuracy score is", accuracy_score(clust2['Medal_Won'], labels))
```

### With the last code, it can be check if there is a possibility to get a higher accuracy score due to the permuted labels, but in the end We obtain the same value, so it's easy to see that 0.77 is the higher value We can obtain for now

## Association rule

```python
d2.rename(columns={'Medal':'Total medals','2016':'Most recent GPA'},inplace=True)
d2.head()
```

```python
colty=[]
for i,j in zip(d2["Agriculture"],d2["Industry"]):
    i = i.replace(',', '.')
    j = j.replace(',', '.')
    if float(i)>=float(j):
        colty.append("agriculture")
    elif float(i)<=float(j):
        colty.append("industry")
d2["Type"]=colty
d2=d2.drop(columns=["Industry","Agriculture"])
d2.head(30)
```

```python
mm=d2["Medal_Won"].mean()
mp=d2["Population"].mean()
mg=d2["Most recent GPA"].mean()
colm=[]
colp=[]
colg=[]
display(mm,mp,mg)
```

```python
for i,j,k in zip(d2["Medal_Won"],d2["Population"],d2["Most recent GPA"]):
    if int(i)<=4:
        a="<= 4 medals"
        colm.append(a)
    elif int(i)>4:
        b=">4 medals"
        colm.append(b)
    if int(j)<=int(mp):
        a="<= 57488090 people"
        colp.append(a)
    elif int(j)>int(mp):
        b=">57488090 people"
        colp.append(b)
    if float(k)<=mg:
        a="<=16239,6 GPA"
        colg.append(a)
    elif float(k)>mg:
        b=">16239,6 GPA"
        colg.append(b)
```

```python
data=pd.DataFrame({"Medal_Won":colm,"Population":colp,"Most recent GPA":colg,"Type":d2["Type"],"Country":d2["Country"]})
data.head(20)
```

```python
d=data.values.tolist()
```

```python
te = TransactionEncoder()
te_ary = te.fit(d).transform(d)
te_ary.astype("int")
```

```python
data2 = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
frequent_itemsets = apriori(data2, min_support = 0.1, use_colnames = True)
frequent_itemsets.head(15)
```

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold = 1)
rules.head()
```
