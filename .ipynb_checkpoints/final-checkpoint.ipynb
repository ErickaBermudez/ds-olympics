#### Final data frame with the principal data of Medals Won per country categorized in:
#### 1 - High Performance
#### 0 - Regular Performance

```python
# find the appropriate cluster number
from sklearn.cluster import KMeans
seed = 4

inertia = []
elbow = dict()
for i in range(1, 11):
    
    estimator = KMeans(n_clusters = i,random_state=seed)
    res = estimator.fit_predict(clust2)
    inertia = estimator.inertia_
    elbow[i] = inertia
    
elbow_df = pd.Series(elbow)
ax = elbow_df.plot(title = 'Elbow Method')
ax.set_xlabel('Number of clusters')
ax.set_ylabel('Inertia')
plt.plot(2,elbow_df[2],'ro')
```

### We aplied the elbow method to find out the number of clusters that We should use for this analisys

```python
from sklearn.cluster import KMeans
model = KMeans(2)
model.fit(clust2)
print(model.cluster_centers_)
print(model.labels_)
```

```python
#plt.scatter(clust2['Agriculture'],clust2['Service'], c=model.labels_);
#plt.scatter(model.cluster_centers_[:,3], model.cluster_centers_[:,4], s=10, color="red"); # Show the centres
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(clust2['Medal'],model.labels_)
print("Accuracy score is", acc)
```

### With the value of 2 clusters for the problem We obtain an accuracy of 0.77, which is  relatively good

```python
clusters = model.fit_predict(clust2)

# Fix permuted labels

from scipy import stats
labels = np.zeros_like(clusters)
for i in range(4):
    mask = (clusters == i)
    labels[mask] = stats.mode(clust2['Medal'][mask])[0]
    
    
print("Accuracy score is", accuracy_score(clust2['Medal'], labels))
```

### With the last code, it can be check if there is a possibility to get a higher accuracy score due to the permuted labels, but in the end We obtain the same value, so it's easy to see that 0.77 is the higher value We can obtain for now

## Association rule

```python
d2.rename(columns={'Medal':'Total medals','2016':'Most recent GPA'},inplace=True)
d2.head()
```

```python
colty=[]
for i,j in zip(d2["Agriculture"],d2["Industry"]):
    i = i.replace(',', '.')
    j = j.replace(',', '.')
    if float(i)>=float(j):
        colty.append("agriculture")
    elif float(i)<=float(j):
        colty.append("industry")
d2["Type"]=colty
d2=d2.drop(columns=["Industry","Agriculture"])
d2.head(30)
```

```python
mm=d2["Total medals"].mean()
mp=d2["Population"].mean()
mg=d2["Most recent GPA"].mean()
colm=[]
colp=[]
colg=[]
display(mm,mp,mg)
```

```python
for i,j,k in zip(d2["Total medals"],d2["Population"],d2["Most recent GPA"]):
    if int(i)<=4:
        a="<= 4 medals"
        colm.append(a)
    elif int(i)>4:
        b=">4 medals"
        colm.append(b)
    if int(j)<=int(mp):
        a="<= 57488090 people"
        colp.append(a)
    elif int(j)>int(mp):
        b=">57488090 people"
        colp.append(b)
    if float(k)<=mg:
        a="<=16239,6 GPA"
        colg.append(a)
    elif float(k)>mg:
        b=">16239,6 GPA"
        colg.append(b)
```

```python
data=pd.DataFrame({"Total medals":colm,"Population":colp,"Most recent GPA":colg,"Type":d2["Type"],"Country":d2["Country"]})
data.head(20)
```

```python
d=data.values.tolist()
```

```python
te = TransactionEncoder()
te_ary = te.fit(d).transform(d)
te_ary.astype("int")
```

```python
data2 = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
frequent_itemsets = apriori(data2, min_support = 0.1, use_colnames = True)
frequent_itemsets.head(15)
```

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold = 1)
rules.head()
```

```python
#top 10 rules
rules.sort_values(['support'], axis=0, ascending=False, inplace=True)
rules.head(10)
```

From here we can see that most important relation is between people being less and the drop in the quantity of medals. The drop of the GPA and the quantity of medals is also highly related. Actualy, those three would be the most important features

```python
rules.tail(10)
```

The least important feature is the agriculture/industry. We also see that the relation affects more for getting low values than high values.

```python
rules[rules['antecedents'] == {'>4 medals'}]
```

There isn't anything significantly related with getting a lot of medals

```python
plt.scatter(rules['support'], rules['confidence'], alpha=0.5)
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
plt.show()
```
